#### Collaberative Filtering Based Recommendation system

```python
# Select user IDs with more than 200 book ratings
min_threshold_ratings = 200

#Count book-ratings per User
num_ratings_per_user = df_merge.groupby('User-ID')['Book-Rating'].count()

# Filter users with more than the minimum threshold
knowledgeable_user_ids = num_ratings_per_user[num_ratings_per_user > min_threshold_ratings].index
```

```python
num_ratings_per_user
```

```python
# Length of the users who have given more than 200 book rating
len(knowledgeable_user_ids)
```

```python
## filter Ratings From knowledgable users
knowledgeable_user_ratings = df_merge[df_merge['User-ID'].isin(knowledgeable_user_ids)]
```

```python
min_rating_count_threshold = 50
rating_count = knowledgeable_user_ratings.groupby('Book-Title').count()['Book-Rating']
popular_books = rating_count[rating_count >= min_rating_count_threshold].index
```

```python
rating_count.head()
```

```python
final_ratings = knowledgeable_user_ratings[knowledgeable_user_ratings['Book-Title'].isin(popular_books)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID',values='Book-Rating')
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
def recommend(Book_Name):
    index = np.where(pt.index==Book_Name)[0][0]
    similar_books = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    
    data =[]
    for i in similar_books:
        item = []
        temp_df = df_books[df_books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-L'].values))

        data.append(item)
        
    return data
    
```

```python
recommend('4 Blondes')
```

```python
recommend('Wish You Well')
```
